```python
def preprocess(df):
    df= df.replace([np.inf, -np.inf], np.nan).dropna(subset=df.columns)
    return df

df = preprocess(df)
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt


X = df[cols]
X = sm.add_constant(X)  # æ·»åŠ å¸¸æ•°é¡¹
y = df['1m']
model = sm.OLS(y, X).fit()

# è®­ç»ƒæ•°æ®é›†çš„é¢„æµ‹å€¼
df['Pred'] = model.predict(X)

# è®­ç»ƒæ•°æ®é›†çš„æ®‹å·®
df['resid'] = model.resid

# ç»˜åˆ¶æ”¶å…¥ä¸Žæ®‹å·®çš„æ•£ç‚¹å›¾
df.plot('Pred', 'resid', kind='scatter')
plt.xlabel('Pred')
plt.ylabel('Residuals')
plt.title('Pred vs Residuals')
plt.show()

```

```python
df['resid_t'] = (df['resid'] - df['resid'].mean()) / df['resid'].std()
```

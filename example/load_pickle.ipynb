```python
import os, sys
sys.path.insert(0, "../")
import datetime
import gzip 
import pickle
import gc
import dolphindb.settings as keys
import warnings
import pandas as pd
pd.set_option("display.max_rows", 200)

from factor_cal.utils.tools import show_memory
```

```python
base_path = "/data2/ddb_data"
print("Loading snap info")
show_memory("  before load_snap_info")
date_str = "20240219"
snap_filepath = os.path.join(base_path, 'ticker_' + date_str + '.pkl')
```

```python
def read_pickle(filepath):
    with gzip.open(filepath, 'rb') as f:
        data = pickle.load(f)
    return data
```

```python
data = read_pickle(snap_filepath)
```

```python
result = {}
total = 0
error_num = 0
for date in pd.date_range("2024-02-07", periods=100):
    date = date.strftime("%Y%m%d")
    snap_filepath = os.path.join(base_path, 'ticker_' + date + '.pkl')
    if os.path.exists(snap_filepath):
        data = read_pickle(snap_filepath)
        show_memory(f"{date}:  after load_snap_info")
        result[date] = []
        data = data[(data['bid_size1'] < 0) | (data['ask_size1'] < 0)]
        result[date].append(data.shape[0])
        result[date].append(data['code'].unique().tolist())
        total += 1
        if data.shape[0] > 0:
            error_num += 1
        del data
        gc.collect()

print('total:', total, 'error_num:', error_num, 'portion:', error_num/total)
```

```python
import pickle
pickle.dump(result, open("pkl/snap_error_data.pkl", "wb"))
```
